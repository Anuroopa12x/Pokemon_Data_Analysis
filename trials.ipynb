```python
# Creating a function for finding stats of a pokemon 
def scrap_pokemon_details(pokemon_name):
    # requesting text
    text = requests.get(f"https://bulbapedia.bulbagarden.net/wiki/{pokemon_name}_(Pok%C3%A9mon)").text
    
    # Creating BeautifulSoup instance
    soup = BeautifulSoup(text, "lxml")
    
    # Hp 
    hp_data = soup.find("tr", style = "background: #9EE865; text-align:center")
    hp_data2 = hp_data.find_all("div")
    hp = hp_data2[1].text.strip()
    if hp.isnumeric():
        hp = int(hp)
    else:
        print("Hp is not numeric")
    
    # attack
    attack_data = soup.find("tr", style = "background: #F5DE69; text-align:center") 
    attack_data2 = attack_data.find_all("div")
    attack = attack_data2[1].text.strip()
    if attack.isnumeric():
        attack = int(attack)
    else:
        print("attack is not numeric")
    
    # defense
    defense_data = soup.find("tr", style = "background: #F09A65; text-align:center")
    defense_data2 = defense_data.find_all("div")
    defense = defense_data2[1].text.strip()
    if defense.isnumeric():
        defense = int(defense)
    else:
        print("Defense is not numeric")
    
    # special attack
    special_attack_data = soup.find("tr", style = "background: #66D8F6; text-align:center")
    special_attack_data2 = special_attack_data.find_all("div")
    special_attack = special_attack_data2[1].text.strip()
    if special_attack.isnumeric():
        special_attack = int(special_attack)
    else:
        print("Special Attack is not numeric")
    
    # special defense
    special_defense_data = soup.find("tr", style = "background: #899EEA; text-align:center")
    special_defense_data2 = special_defense_data.find_all("div")
    special_defense = special_defense_data2[1].text.strip()
    if special_defense.isnumeric():
        special_defense = int(special_defense)
    else:
        print("Special Defense is not numeric")
        
    # speed
    speed_data = soup.find("tr", style="background: #E46CCA; text-align:center")
    speed_data2 = speed_data.find_all("div")
    speed = speed_data2[1].text.strip()
    if speed.isnumeric():
        speed = int(speed)
    else:
        print("Speed is not numeric")
    # total
    total = hp + attack + defense + special_attack + special_defense + speed
    
    return {"name" : pokemon_name, "hp" : hp, "attack" : attack, "defense" : defense,
           "special_attack" : special_attack, "special_defense" : special_defense,
           "speed" : speed, "total" : total}
```

```python
# gen1 starters
bulbasaur_data = scrap_pokemon_details("Bulbasaur")
pokemon.loc[1] = bulbasaur_data
ivysaur_data = scrap_pokemon_details("Ivysaur")
pokemon.loc[2] = ivysaur_data
venusaur_data = scrap_pokemon_details("Venusaur")
pokemon.loc[3] = venusaur_data
```

```python
def get_moves():
    moves = []
    text = requests.get("https://pokemondb.net/move/all").text
    soup = BeautifulSoup(text, "lxml")
    data = soup.find("table", class_="data-table sticky-header block-wide")
    data_moves = data.find_all("tr")
    data_moves = data_moves[1: ]
    
    for dta in data_moves:
        #name
        name = dta.find("a", class_="ent-name").text
        #type
        type_ = dta.find("td", class_="cell-icon").text
        #category
        if dta.find("img", class_="img-fixed") != None:
            category = dta.find("img", class_="img-fixed").get("alt")
        else:
            category = None
        # power, accuracy, probability 
        numericals = dta.find_all("td", class_="cell-num")
        power = numericals[0].text
        accuracy = numericals[1].text
        pp = numericals[2].text
        prob = numericals[3].text

        # effect
        effect = dta.find("td", class_="cell-long-text").text

        moves.append({"name": name, "type": type_, "category": category,
                     "power": power, "accuracy": accuracy, "pp": pp,
                      "effect": effect, "probability%": prob})
    return pd.DataFrame(moves)
```

```python
def get_level_up_moves(pokemon_name):
    pokemon_name = pokemon_name.lower().replace(" ", "")
    # requesting data
    try:
        response = requests.get(f"https://www.serebii.net/pokedex-sv/{pokemon_name}/")
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"A Request error occured {e}")
        return None
    text = response.text
    soup = soup = BeautifulSoup(text, "lxml")
    data = soup.find_all("table", class_="dextable")
    print(data)
    print(len(data))
    #intialising emppty set
    moves = []
    # get moves 
    moves_data = data[8].find_all("td", class_="fooinfo")
    for m in moves_data:
        move = m.find("a")
        if move != None:
            moves.append(move)
    return {"name": pokemon_name, "moves": moves}
```

```python
def get_entire_data():
    text = requests.get("https://pokemondb.net/pokedex/all").text
    soup = BeautifulSoup(text, "lxml")
    return_data = []
    data = soup.find("table", id="pokedex")
    data1 = data.find_all("tr")
    
    for dta in data1[1: ]:
        #number 
        number = dta.find(["td", "span"], class_="infocard-cell-data").text
        #name
        name_data = dta.find("td", class_="cell-name")
        name_data_main = name_data.find("a", class_="ent-name").text
        if name_data.find("small", class_="text-muted") != None:
            name_data_sub = name_data.find("small", class_="text-muted").text
            name = name_data_main + " " + name_data_sub
        else:
            name = name_data_main
        # type
        type_data = dta.find("td", class_="cell-icon")
        dual_not = type_data.find_all("a")
        if len(dual_not) > 1:
            type1 = dual_not[0].text
            type2 = dual_not[1].text
        else:
            type1 = dual_not[0].text
            type2 = "Nil"
        # stats 
        total = dta.find("td", class_="cell-num cell-total").text
        other_stats = dta.find_all("td", class_="cell-num")
        hp = other_stats[2].text.replace("\n", "")
        attack = other_stats[3].text
        defense = other_stats[4].text
        special_attack = other_stats[5].text
        special_defense = other_stats[6].text
        speed = other_stats[7].text
        
        return_data.append({"number": number, "name": name, "type1": type1,
                          "type2": type2, "total": total, "hp": hp,
                          "attack": attack, "defense": defense, "special_attack": special_attack,
                          "special_defense": special_defense, "speed": speed})
    return pd.DataFrame(return_data)
```

```python
def preparing_data(details, level=50):
    details = details.copy()
    stats = ["attack", "special_attack", "defense", "special_defense", "speed", "hp"]
    for stat in stats:
        base = details[stat].astype(int)
        details[stat] = np.floor(((2 * base + 31) * level) / 100) + (level + 10 if stat == "hp" else 5)
    details["main_attack"] = np.maximum(details["attack"], details["special_attack"])
    return details

entire_data = preparing_data(entire_data, 50)
```
