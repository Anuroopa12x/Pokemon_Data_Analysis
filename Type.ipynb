```python
def type_calc(li):
    types = ["fire", "dark", "flying", "fighting", "bug", "fairy", "water", "ice", "electric", "dragon",
        "psychic", "rock", "ground", "poison", "ghost", "normal", "grass", "steel"]
    s = pd.DataFrame(columns = ["type", "total", "regular" "legendary", "paradox", "ultra_beast"], index = range(0, 18))
    for i, type_ in enumerate(types):
        s.loc[i, "type"] = type_
        s.loc[i, "total"] = li[li["type"].str.contains(type_, case = False) == True]["name"].count()
        s.loc[i, "regular"] = li[(li["type"].str.contains(type_, case = False) == True) & 
                                (li["legendary"] == "No") & (li["ability"] != "Beast Boost")&
                                (li["category"] != "Paradox Pokémon")]["name"].count()
        s.loc[i, "legendary"] = li[(li["type"].str.contains(type_, case = False) == True) & 
                                (li["legendary"] == "Yes")]["name"].count()
        s.loc[i, "paradox"] = li[(li["category"] == "Paradox Pokémon") & 
                                 (li["type"].str.contains(type_, case = False) == True)]["name"].count()
        s.loc[i, "ultra_beast"] = li[(li["type"].str.contains(type_, case = False) == True) & 
                                (li["ability"] == "Beast Boost")]["name"].count()
        
    return s


type_dict = type_calc(details)
type_dict.drop(columns = "regularlegendary", inplace = True)
type_dict
```

```python
def gen_type(li):
    s = pd.DataFrame()
    s["generation"] = range(1, 10)
    types = ["fire", "dark", "flying", "fighting", "bug", "fairy", "water", "ice", "electric", "dragon",
        "psychic", "rock", "ground", "poison", "ghost", "normal", "grass", "steel"]
    for type_ in types:
        type_counts = li[li["type"].str.contains(type_, case = False)]["gen"].value_counts(sort = False)
        
        type_df = pd.DataFrame({type_ : type_counts}).reset_index()
        
        type_df.rename(columns = {"index" : "generation"}, inplace = True)
        
        s = s.merge(type_df, on = "generation", how = "left")
    return s.fillna(0)
```

```python
import pandas as pd
import numpy as np

def calculate_damage(level, base_power, attack, defense, type_damage):
    """Calculates the damage dealt in a single hit. This is a vectorized function."""
    a_d = attack / defense
    damage = ((((2 * level) / 5 + 2) * base_power * a_d) / 50 + 2) * type_damage
    return damage

def get_type_damage(type_chart, attacking_type, defending_type1, defending_type2):
    """Calculates the type effectiveness multiplier. This is a vectorized function."""
    # Ensure all inputs are arrays for vectorized lookup
    defending_type1_arr = defending_type1.values if isinstance(defending_type1, pd.Series) else np.array(defending_type1)
    defending_type2_arr = defending_type2.values if isinstance(defending_type2, pd.Series) else np.array(defending_type2)
    
    type_damage1 = type_chart.loc[attacking_type, defending_type1_arr].values
    
    # Create an array for the second type's damage.
    # It defaults to 1.0 (no effect) for single-type Pokemon ('None').
    # Otherwise, it looks up the actual damage multiplier.
    type_damage2 = np.ones_like(defending_type2_arr, dtype=float)
    valid_type2_mask = (defending_type2_arr != "None")
    if np.any(valid_type2_mask):
        type_damage2[valid_type2_mask] = type_chart.loc[attacking_type, defending_type2_arr[valid_type2_mask]].values
    
    combined_damage = type_damage1 * type_damage2
    return combined_damage

def type_effectiveness_fast(li, type_chart, i, j):
    """
    Finds which type prevails over another using a highly-optimized, vectorized approach.
    Parameters: i, j - any 2 types, li - a dataframe, type_chart - a dataframe containing type matchups.
    Returns: A dict with wins and loses of one type to another.
    """
    # Filter the data for each type in one step
    i_data = li[li["type"].str.contains(i, case=False)].reset_index(drop=True)
    j_data = li[li["type"].str.contains(j, case=False)].reset_index(drop=True)
    
    # Add type1 and type2 columns
    for df in [i_data, j_data]:
        df[['type1', 'type2']] = df['type'].str.split(' / ', expand=True).fillna('None')

    # Create a cross-join DataFrame of all possible battles
    i_data['key'] = 1
    j_data['key'] = 1
    
    battle_df = pd.merge(i_data, j_data, on='key', suffixes=('_i', '_j'), how='outer').drop(columns='key')

    # Create a copy to modify and set the index
    type_chart_indexed = type_chart.set_index('Type')
    
    level = 50
    base_power = 50

    # Vectorized damage and stat calculations
    battle_df['attack_i'] = battle_df[['attack_i', 'special_attack_i']].max(axis=1)
    battle_df['defense_j'] = np.where(battle_df['attack_i'] >= battle_df['special_attack_i'], 
                                       battle_df['defense_j'], 
                                       battle_df['special_defense_j'])

    battle_df['attack_j'] = battle_df[['attack_j', 'special_attack_j']].max(axis=1)
    battle_df['defense_i'] = np.where(battle_df['attack_j'] >= battle_df['special_attack_j'], 
                                       battle_df['defense_i'], 
                                       battle_df['special_defense_i'])

    # Vectorized type effectiveness calculation
    battle_df['damage_multiplier_i'] = get_type_damage(type_chart_indexed, i, battle_df['type1_j'], battle_df['type2_j'])
    battle_df['damage_multiplier_j'] = get_type_damage(type_chart_indexed, j, battle_df['type1_i'], battle_df['type2_i'])

    # Vectorized total damage calculation
    battle_df['damage_i_to_j'] = calculate_damage(level, base_power, battle_df['attack_i'], battle_df['defense_j'], battle_df['damage_multiplier_i'])
    battle_df['damage_j_to_i'] = calculate_damage(level, base_power, battle_df['attack_j'], battle_df['defense_i'], battle_df['damage_multiplier_j'])
    
    # Calculate hits to knockout (KO) for each Pokémon
    hp_i = battle_df['hp_i'] + 100
    hp_j = battle_df['hp_j'] + 100
    
    battle_df['hits_i_to_ko'] = np.ceil(hp_j / battle_df['damage_i_to_j'])
    battle_df['hits_j_to_ko'] = np.ceil(hp_i / battle_df['damage_j_to_i'])
    
    # Determine the winner based on hits to KO and speed
    # Start with the assumption that the faster Pokémon wins if hits are equal
    battle_df['winner'] = np.where(battle_df['hits_i_to_ko'] <= battle_df['hits_j_to_ko'], 1, -1)
    
    # If hits are equal, the faster Pokémon wins
    equal_hits_mask = battle_df['hits_i_to_ko'] == battle_df['hits_j_to_ko']
    battle_df.loc[equal_hits_mask, 'winner'] = np.where(battle_df.loc[equal_hits_mask, 'speed_i'] >= battle_df.loc[equal_hits_mask, 'speed_j'], 1, -1)

    # Count the wins for each type
    streak = {
        "win_i": (battle_df['winner'] == 1).sum(),
        "win_j": (battle_df['winner'] == -1).sum()
    }

    return streak
```
